```python
# 📌 Step 1: 라이브러리 설치 (처음 1회만 실행)
# !pip install yfinance pandas

# 📌 Step 2: 라이브러리 임포트
import yfinance as yf
import pandas as pd
import time
from datetime import datetime
import os
```

```python
# 📌 Step 3: 종목 리스트 정의
tickers = {
    "Tesla": "TSLA", "Apple": "AAPL", "Nvidia": "NVDA", "Microsoft": "MSFT",
    "Amazon": "AMZN", "AMD": "AMD", "Meta": "META", "Alphabet": "GOOGL",
    "SPY ETF": "SPY", "QQQ ETF": "QQQ",
    "KOSPI": "^KS11", "KOSDAQ": "^KQ11",
    "삼성전자": "005930.KQ", "SK하이닉스": "000660.KQ", "현대차": "005380.KQ",
    "네이버": "035420.KQ", "카카오": "035720.KQ", "LG에너지솔루션": "373220.KQ",
    "삼성SDI": "006400.KQ", "POSCO홀딩스": "005490.KQ", "KB금융": "105560.KQ", "셀트리온": "068270.KQ"
}
```

```python
# 📌 Step 4: 기간 설정
start_date = "2020-01-01"
end_date = datetime.today().strftime("%Y-%m-%d")

# 📌 Step 5: 저장 디렉토리 생성
save_dir = "yfinance_data_5y"
os.makedirs(save_dir, exist_ok=True)

# 📌 Step 6: 데이터 수집 + 저장
collected = {}
for name, symbol in tickers.items():
    try:
        print(f"📥 {name} ({symbol}) 수집 중...")
        df = yf.download(symbol, start=start_date, end=end_date, progress=False)
        if not df.empty:
            df["Name"] = name
            df["Symbol"] = symbol
            collected[name] = df
            df.to_csv(f"{save_dir}/{name}_{symbol}.csv", encoding="cp949")
        else:
            print(f"⚠️ {name} 데이터가 비어 있음")
    except Exception as e:
        print(f"❌ {name} 오류: {e}")
    time.sleep(2)

```

```python
# 📌 Step 7: 통합 파일 저장
all_data = pd.concat(collected.values())
all_data.reset_index(inplace=True)
all_data.to_csv(f"{save_dir}/all_stocks_5y_combined.csv", index=False)
all_data.head()
```
