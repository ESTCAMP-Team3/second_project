```python
# 통합 리스트
symbols = overseas + domestic
```

```python
# 심볼 ↔ 한글 이름 수동 매핑
kor_name_map = {
    "AAPL": "애플",
    "TSLA": "테슬라",
    "MSFT": "마이크로소프트",
    "NVDA": "엔비디아",
    "AMZN": "아마존",
    "META": "메타",
    "AMD": "AMD",
    "GOOGL": "알파벳",
    "SPY": "S&P500 ETF",
    "QQQ": "나스닥 ETF",

    "005930.KS": "삼성전자",
    "000660.KS": "SK하이닉스",
    "005380.KS": "현대차",
    "035420.KQ": "네이버",
    "035720.KQ": "카카오",
    "373220.KS": "LG에너지솔루션",
    "006400.KS": "삼성SDI",
    "005490.KS": "POSCO홀딩스",
    "105560.KS": "KB금융",
    "068270.KQ": "셀트리온",
}
```

```python
# 결과 저장용 리스트
result = []
```

```python
for sym in symbols:
    try:
        ticker = yf.Ticker(sym)
        info = ticker.info
        name = info.get("longName", "N/A")
        kor_name = kor_name_map.get(sym, "N/A")
        result.append({
            "Symbol": sym,
            "Company Name": name,
            "Korean Name": kor_name
        })
    except Exception as e:
        result.append({
            "Symbol": sym,
            "Company Name": f"Error: {e}",
            "Korean Name": kor_name_map.get(sym, "N/A")
        })
```

```python
# DataFrame 생성 및 출력
df_map = pd.DataFrame(result)
df_map
```

```python
# 3. 결과 저장 구조
data_overseas = {}
data_domestic = {}
map_overseas = []
map_domestic = []
```

```python
# 1. 수동 매핑: Symbol ↔ 한글 이름
kor_name_map = {
    "AAPL": "애플", "TSLA": "테슬라", "MSFT": "마이크로소프트",
    "NVDA": "엔비디아", "AMZN": "아마존", "META": "메타",
    "AMD": "AMD", "GOOGL": "알파벳", "SPY": "S&P500 ETF", "QQQ": "나스닥 ETF",
    "005930.KS": "삼성전자", "000660.KS": "SK하이닉스", "005380.KS": "현대차",
    "035420.KQ": "네이버", "035720.KQ": "카카오", "373220.KS": "LG에너지솔루션",
    "006400.KS": "삼성SDI", "005490.KS": "POSCO홀딩스", "105560.KS": "KB금융", "068270.KQ": "셀트리온"
}

# 2. 심볼 리스트 나누기
overseas = [sym for sym in kor_name_map if ".KS" not in sym and ".KQ" not in sym]
domestic = [sym for sym in kor_name_map if ".KS" in sym or ".KQ" in sym]


# 4. 결과 저장 딕셔너리
data_overseas = {}
data_domestic = {}

# 5. 수집 함수
def collect_data(symbols, target_dict, label):
    for sym in symbols:
        try:
            df = yf.Ticker(sym).history(period='5y')
            if not df.empty:
                name = kor_name_map[sym]
                target_dict[name] = df
                print(f"[✓] {label}: {name} 수집 완료")
            else:
                print(f"[X] {label}: {sym} → 데이터 없음")
        except Exception as e:
            print(f"[!] {label}: {sym} 실패: {e}")

# 6. 실행
collect_data(overseas, data_overseas, "해외")
collect_data(domestic, data_domestic, "국내")

print("\n✅ 완료: 해외 {}종, 국내 {}종 수집됨".format(len(data_overseas), len(data_domestic)))
```

# 암호화폐 데이터

```python
import ccxt
import datetime

# 바이낸스 객체 생성
binance = ccxt.binance()

# fetch_ohlcv로 비트코인 일봉 데이터 수집
symbol = 'BTC/USDT'
timeframe = '1d'
limit = 1000  # 약 3년치

ohlcv = binance.fetch_ohlcv(symbol, timeframe=timeframe, limit=limit)

# DataFrame으로 변환
df_coin = pd.DataFrame(ohlcv, columns=['timestamp', 'open', 'high', 'low', 'close', 'volume'])
df_coin['timestamp'] = pd.to_datetime(df_coin['timestamp'], unit='ms')

df_coin.head(14)
```

```python
df_coin.to_csv('/home/nshawny/AI project 2/personal_data/coin/coin.csv')
```
