```python
# 📌 라이브러리
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from sklearn.metrics import mean_squared_error, r2_score

# 📌 1. 데이터 수집
aapl = yf.download('AAPL', start='2017-01-01', end='2022-12-31')
aapl = aapl[['Open','High','Low','Close','Volume']].dropna()
aapl.reset_index(drop=True, inplace=True)

# 📌 2. 타겟 컬럼 추가
aapl['Close_Target'] = aapl['Close']

# 📌 3. 날짜 인덱스 추출
dates = yf.download('AAPL', start='2017-01-01', end='2022-12-31').index
dates = dates[20:]

# 📌 4. 2022년 1월 기준 Split
split_date = '2022-01-03'
split_idx = np.where(dates >= np.datetime64(split_date))[0][0]
print('2022년 분할 위치:', split_idx)

# 📌 5. 분할
train_df = aapl.iloc[:split_idx+20].copy()
test_df = aapl.iloc[split_idx+20:].copy()
print('Train:', train_df.shape)
print('Test:', test_df.shape)

# ✅ 6. 스케일링 누수 방지
scaler = StandardScaler()
train_features = scaler.fit_transform(train_df.iloc[:, :5])
test_features = scaler.transform(test_df.iloc[:, :5])

# 📌 7. 스케일링 적용 데이터프레임 생성
train_df_scaled = pd.DataFrame(train_features, columns=['Open','High','Low','Close','Volume'])
train_df_scaled['Close_Target'] = train_df['Close_Target'].values

test_df_scaled = pd.DataFrame(test_features, columns=['Open','High','Low','Close','Volume'])
test_df_scaled['Close_Target'] = test_df['Close_Target'].values

# 📌 8. 윈도우 생성 함수
def create_sequences(df, window):
    x_list, y_list = [], []
    for i in range(len(df) - window):
        x_tmp = df.iloc[i:i+window, 0:5].values
        y_tmp = df.iloc[i+window, 5]
        x_list.append(x_tmp)
        y_list.append(y_tmp)
    return np.array(x_list), np.array(y_list)

윈사이즈 = 20
x_train, y_train = create_sequences(train_df_scaled, 윈사이즈)
x_test, y_test = create_sequences(test_df_scaled, 윈사이즈)

print('Train:', x_train.shape, y_train.shape)
print('Test:', x_test.shape, y_test.shape)

# 📌 9. 모델 설계
model = Sequential([
    LSTM(48, activation='tanh', input_shape=(윈사이즈, 5)),
    Dropout(0.4),
    Dense(64, activation='relu'),
    Dropout(0.4),
    Dense(1)
])
model.compile(loss='mae', optimizer='adam', metrics=['mae'])

# 📌 10. 콜백
cbk = ModelCheckpoint('best_model.h5', save_weights_only=True, monitor='val_mae', save_best_only=True)
es = EarlyStopping(patience=20, restore_best_weights=True)

# 📌 11. 학습
history = model.fit(
    x_train, y_train,
    validation_split=0.2,
    epochs=100,
    batch_size=32,
    callbacks=[cbk, es],
    verbose=2
)

# 📌 12. 평가
model.load_weights('best_model.h5')
test_loss = model.evaluate(x_test, y_test, verbose=0)
print('📌 Test MAE:', test_loss[0])

# 📌 13. 예측
y_pred = model.predict(x_test)

# 📌 14. 지표 계산
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)
print(f"✅ Test RMSE: {rmse:.2f}")
print(f"✅ Test R²: {r2:.4f}")


```

```python
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
def evaluate_metrics(y_true, y_pred):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100

    print(f"✅ RMSE: {rmse:.4f}")
    print(f"✅ MAE: {mae:.4f}")
    print(f"✅ R²: {r2:.4f}")
    print(f"✅ MAPE: {mape:.2f}%")
    
    return {'RMSE': rmse, 'MAE': mae, 'R2': r2, 'MAPE': mape}
```
